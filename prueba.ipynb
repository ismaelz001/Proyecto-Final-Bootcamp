```python
import pandas as pd

# Cargar los archivos CSV
df1   # Contiene 'producto_id' y 'url'
df2  # Contiene 'url' y características

# Verificar que las columnas claves existen
if 'url' not in df1.columns or 'url' not in df2.columns:
    raise ValueError("Ambos archivos deben contener la columna 'url'.")

# Unir los DataFrames en base a la URL
df_caracteristicas = df2.merge(df1[['producto_id', 'url']], on='url', how='left')

# Mover 'producto_id' al inicio
columnas = ['producto_id'] + [col for col in df_caracteristicas.columns if col not in ['producto_id', 'url']]
df_caracteristicas = df_caracteristicas[columnas]

# Eliminar la columna 'url' porque ya no se necesita
#df_caracteristicas.drop(columns=['url'], inplace=True)

# Guardar el nuevo CSV listo para la base de datos
df_caracteristicas.to_csv("data/caracteristicas_portatiles_actualizado.csv", index=False)

print("Archivo 'caracteristicas_portatiles_actualizado.csv' generado correctamente con producto_id.")
```
