```python
df_caracteristicas['Weight'] = df_caracteristicas['Weight'].astype(str)  
df_caracteristicas['Weight'] = df_caracteristicas['Weight'].str.extract(r'([\d.]+)')
df_caracteristicas['Weight'] = pd.to_numeric(df_caracteristicas['Weight'], errors='coerce')
df_caracteristicas['Weight'] = df_caracteristicas['Weight'].astype(float)  
df_caracteristicas['Battery mAh'] = df_caracteristicas['Battery mAh'].astype(str)  
df_caracteristicas['Battery mAh'] = df_caracteristicas['Battery mAh'].str.extract(r'([\d.]+)')
df_caracteristicas['Battery mAh'] = pd.to_numeric(df_caracteristicas['Battery mAh'], errors='coerce')
df_caracteristicas['Battery mAh'] = df_caracteristicas['Battery mAh'].astype(float) 
```

```python
df_merged= pd.read_csv('../productos_portatiles_actualizado.csv')
df_merged.head()
```

```python
df_merged['opiniones'] = df_merged['opiniones'].astype(str)
df_merged['opiniones'] = df_merged['opiniones'].str.replace(r'\D', '', regex=True)
df_merged['opiniones'] = df_merged['opiniones'].fillna(0)
df_merged['opiniones'] = df_merged['opiniones'].astype(int)
df_merged.head()
```

```python
df_merged['rating'] = df_merged['rating'].astype(str)
df_merged['rating'] = df_merged['rating'].str.replace(',', '.', regex=False)
df_merged['rating'] = df_merged['rating'].str.replace(r'/5', '', regex=True)
df_merged['rating'] = df_merged['rating'].astype(float)
df_merged.head()
```

```python
# Convertimos todo a string para hacer reemplazos
df_merged['precio_tachado'] = df_merged['precio_tachado'].astype(str)

# Reemplazamos separadores de miles (punto) y cambiamos coma por punto (formato europeo a float)
df_merged['precio_tachado'] = df_merged['precio_tachado'].str.replace('.', '', regex=False)
df_merged['precio_tachado'] = df_merged['precio_tachado'].str.replace(',', '.', regex=False)

# Eliminamos el símbolo de euro si está presente
df_merged['precio_tachado'] = df_merged['precio_tachado'].str.replace(r'€', '', regex=True)

# Rellenamos valores NaN con la columna 'precio'
df_merged['precio_tachado'] = df_merged['precio_tachado'].fillna(df_merged['precio'])

# Convertimos a float después de limpiar los datos
df_merged['precio_tachado'] = df_merged['precio_tachado'].astype(float)

# Si el precio tachado es 0, lo reemplazamos por 'precio'
df_merged['precio_tachado'] = df_merged.apply(
    lambda row: row['precio'] if row['precio_tachado'] == 0 else row['precio_tachado'], axis=1
)

df_merged.head(15)  # Muestra las primeras 15 filas
```

```python
df_merged['timestamp'] = pd.to_datetime(df_merged['timestamp'])
```

```python
# Asegurar que las columnas sean numéricas antes de calcular
df_merged['precio_tachado'] = pd.to_numeric(df_merged['precio_tachado'], errors='coerce')
df_merged['precio'] = pd.to_numeric(df_merged['precio'], errors='coerce')

# Evitar la división por 0: Si precio_tachado == 0, no se calcula el descuento
df_merged['descuento_%'] = df_merged.apply(
    lambda row: ((row['precio_tachado'] - row['precio']) / row['precio_tachado']) * 100
    if row['precio_tachado'] > 0 else 0, axis=1
)

# Redondear a 2 decimales
df_merged['descuento_%'] = df_merged['descuento_%'].round(2)

# Reemplazar valores NaN por 0%
df_merged['descuento_%'] = df_merged['descuento_%'].fillna(0)

# Mostrar primeras 50 filas
df_merged.head(50)
```

```python
marcas = ["Apple", "HP", "Lenovo", "ASUS", "MSI", "Acer", "PcCom", "Alurin"]

# Función para detectar la marca desde el nombre del producto
def detectar_marca(texto):
    for marca in marcas:
        if marca.lower() in texto.lower():
            return marca
    return "Desconocida"
df_merged["marca"] = df_merged["nombre"].apply(detectar_marca)
df_merged.head()
```
