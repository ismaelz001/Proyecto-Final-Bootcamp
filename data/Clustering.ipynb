```python
import pandas as pd
import numpy as np
```

```python
df_productos = pd.read_csv('productos_portatiles_actualizado.csv')
df_productos.head()
```

```python
df_caracteristicas = pd.read_csv('caracteristicas_portatiles_actualizado.csv')
df_caracteristicas.head()
```

```python
df_categorias = pd.read_csv('categoriasPortatiles.csv')
df_categorias.head()
```

```python
#Merge de productos y caracteristicas
df_productos_caracteristicas = pd.merge(df_productos, df_caracteristicas, on='producto_id', how= 'outer')
df_productos_caracteristicas.head()
```

```python
df_merged = pd.merge(df_productos_caracteristicas, df_categorias, left_on='categoria_id', right_on='id', how= 'left')
df_merged.head()
```

```python
df_merged.drop(columns=['id','url_x','url_y','producto_id','categoria_id'], inplace=True)
```

```python
df_merged.head()
```

```python
df_merged.rename(columns={'nombre_y': 'categoria'}, inplace=True)
df_merged.head()
```

```python
df_merged.info()   
```

```python
df_merged['opiniones'] = df_merged['opiniones'].str.replace(r'\D', '', regex=True)
df_merged['opiniones'] = df_merged['opiniones'].fillna(0)
df_merged['opiniones'] = df_merged['opiniones'].astype(int)
df_merged.head()
```

```python
df_merged['rating'] = df_merged['rating'].str.replace(',', '.', regex=False)
df_merged['rating'] = df_merged['rating'].str.replace(r'/5', '', regex=True)
df_merged['rating'] = df_merged['rating'].astype(float)
df_merged.head()
```

```python
df_merged['precio'] = df_merged['precio'].str.replace('.', '', regex=False)
df_merged['precio'] = df_merged['precio'].str.replace(',', '.', regex=False)
df_merged['precio'] = df_merged['precio'].str.replace(r'€', '', regex=True)
df_merged['precio'] = df_merged['precio'].astype(float)
df_merged.head()
```

```python
df_merged['precio_tachado'] = df_merged['precio_tachado'].str.replace('.', '', regex=False)
df_merged['precio_tachado'] = df_merged['precio_tachado'].str.replace(',', '.', regex=False)
df_merged['precio_tachado'] = df_merged['precio_tachado'].str.replace(r'€', '', regex=True)
# df_merged['precio_tachado'] = df_merged['precio_tachado'].fillna(df['precio'])
df_merged['precio_tachado'] = df_merged['precio_tachado'].astype(float)
df_merged.head()
```

```python
df_merged['timestamp'] = pd.to_datetime(df_merged['timestamp'])
```

```python
df_merged['Processor Speed'] = df_merged['Processor Speed'].str.extract(r'([\d.]+)')
df_merged['Processor Speed'] = df_merged['Processor Speed'].astype(float)
df_merged.head()
```

```python
df_merged['Processor Cores'] = df_merged['Processor Cores'].astype(str)
df_merged['Processor Cores'] = df_merged['Processor Cores'].str.extract(r'(\d+)')
df_merged['Processor Cores'] = pd.to_numeric(df_merged['Processor Cores'], errors='coerce')
df_merged['Processor Cores'].astype(float)
df_merged['Processor Cores'].unique()
```

```python
df_merged['RAM Gbs'].unique()
df_merged['RAM Gbs'] = df_merged['RAM Gbs'].str.extract(r'(\d+)')
df_merged['RAM Gbs'] = pd.to_numeric(df_merged['RAM Gbs'], errors='coerce')
df_merged['RAM Gbs'].unique()
```

```python
df_merged['Storage Gbs'] = df_merged['Storage Gbs'].astype(str)
df_merged['Storage Gbs'] = df_merged['Storage Gbs'].str.replace(r'(\d+)\s*TB', lambda x: str(int(x.group(1)) * 1000), regex=True)
df_merged['Storage Gbs'] = df_merged['Storage Gbs'].astype(str)
df_merged['Storage Gbs'] = df_merged['Storage Gbs'].str.extract(r'(\d+)')
df_merged['Storage Gbs'] = pd.to_numeric(df_merged['Storage Gbs'], errors='coerce')
df_merged['Storage Gbs'].astype(float)
df_merged.head()
```

```python
df_merged['Storage Gbs'].unique()
```

```python
df_merged['Display Inches'] = df_merged['Display Inches'].astype(str)  
df_merged['Display Inches'] = df_merged['Display Inches'].str.extract(r'([\d.]+)')
df_merged['Display Inches'] = pd.to_numeric(df_merged['Display Inches'], errors='coerce')
df_merged['Display Inches'].unique()
```

```python
gpu_brands = ['NVidia', 'Intel', 'AMD', 'Apple']

def extract_brand(gpu_name):
    for brand in gpu_brands:
        if brand.lower() in str(gpu_name).lower(): 
            return brand
    return np.nan  

df_merged['GPU Model'] = df_merged['GPU Model'].apply(extract_brand)
df_merged['GPU Model'].unique()
```

```python
df_merged.head()
```

```python
df_merged['USB Ports'] = df_merged['USB Ports'].astype(str).str.extract(r'(\d+)')
df_merged['USB Ports'] = pd.to_numeric(df_merged['USB Ports'], errors='coerce')
df_merged['USB Ports'].unique()
```
