```python
import pandas as pd
```

```python
df = pd.read_csv('productosPortatil.csv')
df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
df["nombre"] = df["nombre"].astype(str)
df["url"] = df["url"].astype(str)
df["categoria"] = df["categoria"].astype(str)

```

```python
df['opiniones'] = df['opiniones'].str.replace(r'\D', '', regex=True)
df['opiniones'] = df['opiniones'].fillna(0)
df['opiniones'] = df['opiniones'].astype(int)
df.head()
```

```python
df['rating'] = df['rating'].str.replace(',', '.', regex=False)
df['rating'] = df['rating'].str.replace(r'/5', '', regex=True)
df['rating'] = df['rating'].astype(float)
```

```python
df['precio'] = df['precio'].str.replace('.', '', regex=False)
df['precio'] = df['precio'].str.replace(',', '.', regex=False)
df['precio'] = df['precio'].str.replace(r'€', '', regex=True)
df['precio'] = df['precio'].astype(float)
```

```python
df.head()
```

```python
df['precio_tachado'] = df['precio_tachado'].str.replace('.', '', regex=False)
df['precio_tachado'] = df['precio_tachado'].str.replace(',', '.', regex=False)
df['precio_tachado'] = df['precio_tachado'].str.replace(r'€', '', regex=True)
# df['precio_tachado'] = df['precio_tachado'].fillna(df['precio'])
df['precio_tachado'] = df['precio_tachado'].astype(float)
df.head()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.head()
```

```python
df.to_csv('productosPortatil-limpio.csv', index=False)
```

```python
df2 = pd.read_csv('../data/caracteristicas_productos_componentes_pc.csv')
df2.head()
```

```python
df2.info()
```

```python
column_names = {
    "Característica_1": "detalle_unico",
    "Característica_2": "pantalla",
    "Característica_3": "bomba",
    "Característica_4": "refrigeracion",
    "Característica_5": "instalacion",
    "Característica_6": "tubos_reforzados",
    "Característica_7": "compatibilidad_procesador",
    }

df2.rename(columns=column_names, inplace=True)
   
df2.head()
```

```python
df2['detalle_unico'] = df2['detalle_unico'].str.replace(r'PERSONALIZACIÓN QUE DESTACA:', '', regex=True)
df2['pantalla'] = df2['pantalla'].str.replace(r'PANTALLA LCD CON GRAN ÁNGULO DE VISIÓN:', '', regex=True)
df2['bomba'] = df2['bomba'].str.replace(r'\BOMBA DE ALTO RENDIMIENTO:', '', regex=True)
df2['refrigeracion'] = df2['refrigeracion'].str.replace(r'REFRIGERACIÓN COLORIDA:', '', regex=True)
df2['instalacion'] = df2['instalacion'].str.replace(r'FÁCIL DE INSTALAR:', '', regex=True)
df2['tubos_reforzados'] = df2['tubos_reforzados'].str.replace(r'TUBOS REFORZADOS:', '', regex=True)
df2['compatibilidad_procesador'] = df2['compatibilidad_procesador'].str.replace(r'COMPATIBILIDAD DE PROCESADOR:', '', regex=True)

df2.head()
```

```python
df2 = df2.drop(['Característica_8','Característica_9','Característica_10', 
               'Característica_11','Característica_12','Característica_13',
               'Característica_14','Característica_15','Característica_16'], axis=1)

df2.head()
```
